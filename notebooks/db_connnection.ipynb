```python
import yaml
import mysql.connector
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('fivethirtyeight') #https://matplotlib.org/stable/gallery/style_sheets/fivethirtyeight.html




import warnings
warnings.filterwarnings('ignore')
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
```

```python
# Load the YAML file
with open("D:\Telangana_Growth_Analysis\db_config.yaml", 'r') as file:
    db_config = yaml.safe_load(file)

# Connect to the database
connection = mysql.connector.connect(**db_config)
cursor = connection.cursor()
```

```python
# Load data from the 'vehicles' table
query_vehicles = "SELECT * FROM vehicles"
vehicles = pd.read_sql(query_vehicles, connection)

# Load data from the 'districts' table
query_districts = "SELECT * FROM districts"
districts = pd.read_sql(query_districts, connection)

# Load data from the 'documents' table
query_documents = "SELECT * FROM documents"
documents = pd.read_sql(query_documents, connection)

# Load data from the 'dim_date' table
query_dim_date = "SELECT * FROM dim_date"
dim_date = pd.read_sql(query_dim_date, connection)

# Load data from the 'investments' table
query_investments = "SELECT * FROM investments"
investments = pd.read_sql(query_investments, connection)

# Display the first few rows of each DataFrame (optional)
print("Vehicles Table:")
print(vehicles.head())
print("\nDistricts Table:")
print(districts.head())
print("\nDocuments Table:")
print(documents.head())
print("\nDim Date Table:")
print(dim_date.head())
print("\nInvestments Table:")
print(investments.head())
```
